```python
from matplotlib import pyplot as plt
import ipywidgets as widgets
from src.tests.time_tester import perform_test_on_graph
```

```python
int_input_0 = widgets.IntText(value=1, description="n", disabled=False)
int_input_1 = widgets.IntText(value=1, description="i", disabled=False)
btn_input_0 = widgets.Button(description="Update", style=widgets.ButtonStyle())
progress_bar = widgets.FloatProgress(value=0.0, description="loading:", min=0.0, max=100.0)
graph_widget = widgets.
def reload(i):
    print(i)
    progress = 0
    plt.title("Pathfinding times")
    plt.xlabel("Number of nodes")
    plt.ylabel("Time in ms")
    leg = plt.legend(loc="upper center")
    runs = {}
    for n in range(2, int_input_0.value):
        for key, value in perform_test_on_graph(number_of_runs=int_input_1.value, number_of_nodes=n).items():
            runs.setdefault(key, [])
            runs[key].append(sum(list(map(lambda x: x.total_time, value))) / len(value))
        progress_bar.value = (n / int_input_0.value) * 100
        progress_bar.description = f"{n} / {int_input_0.value}"
    progress_bar.value = 100
    progress_bar.description = f"progress"
    for key, value in runs.items():
        plt.plot([i / 1000 for i in value], label=key)

btn_input_0.on_click(reload)
widgets.Box([int_input_0, int_input_1, btn_input_0, progress_bar])
```
